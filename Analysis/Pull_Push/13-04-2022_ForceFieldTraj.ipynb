```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
cm = 1/2.54
import glob
```

```python
f0_0l = glob.glob('data/5sec/F0/TipXYF0.0t5S*.txt')
f0_25l = glob.glob('data/5sec/F+0.25/TipXYF+0.25t5S*.txt')
f0_5l = glob.glob('data/5sec/F+0.5/TipXYF+0.5t5S*.txt')
f1_0l = glob.glob('data/5sec/F+1.0/TipXYF+1.0t5S*.txt')
# f1_5l = glob.glob('data/tip/TipXYF1.5**.txt')
f_0_25l = glob.glob('data/5sec/F-0.25/TipXYF-0.25t5S*.txt')
f_0_5l = glob.glob('data/5sec/F-0.5/TipXYF-0.5t5S*.txt')
f_1_0l = glob.glob('data/5sec/F-1.0/TipXYF-1.0t5S*.txt')
```

```python
f0_0 = []
for i in f0_0l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True); _ = _[0::5]
    f0_0.append(_)

f0_25 = []
for i in f0_25l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True); _ = _[0::5]
    f0_25.append(_)

f0_5 = []
for i in f0_5l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True); _ = _[0::5]
    f0_5.append(_)
    
f1_0 = []
for j in f1_0l:
    _ = pd.read_csv(j, names=['t','x','y'], delim_whitespace=True); _ = _[0::5]
    f1_0.append(_)    
    
# f1_5 = []
# for k in f1_5l:
#     _ = pd.read_csv(k, names=['t','x','y'], delim_whitespace=True)
#     f1_5.append(_)
    
f_0_5 = []
for l in f_0_5l:
    _ = pd.read_csv(l, names=['t','x','y'], delim_whitespace=True); _ = _[0::5]
    f_0_5.append(_)
    
f_0_25 = []
for i in f_0_25l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True); _ = _[0::5]
    f_0_25.append(_)
    
f_1_0 = []
for m in f_1_0l:
    _ = pd.read_csv(m, names=['t','x','y'], delim_whitespace=True); _ = _[0::5]
    f_1_0.append(_)
```

```python
f0_0 = f0_0; f0_25 = f0_25; f0_5 = f0_5; f1_0 = f1_0 # default (+ve pull)
#f0_0 = f0_0; f0_25 = f_0_25; f0_5 = f_0_5; f1_0 = f_1_0 # (-ve pull)
```

```python
xdif0_0 = []; ydif0_0 = []
for i in range(len(f0_0)):
    x_ = np.diff(f0_0[i]['x'])
    y_ = np.diff(f0_0[i]['y'])
    xdif0_0.append(x_)
    ydif0_0.append(y_)
    
xdif0_25 = []; ydif0_25 = []
for i in range(len(f0_25)):
    x_ = np.diff(f0_25[i]['x'])
    y_ = np.diff(f0_25[i]['y'])
    xdif0_25.append(x_)
    ydif0_25.append(y_)
    
    
xdif0_5 = []; ydif0_5 = []
for i in range(len(f0_5)):
    x_ = np.diff(f0_5[i]['x'])
    y_ = np.diff(f0_5[i]['y'])
    xdif0_5.append(x_)
    ydif0_5.append(y_)
    
    
xdif1_0 = []; ydif1_0 = []
for i in range(len(f1_0)):
    x_ = np.diff(f1_0[i]['x'])
    y_ = np.diff(f1_0[i]['y'])
    xdif1_0.append(x_)
    ydif1_0.append(y_)
```

```python
ang0_0 = []
for i in range(len(ydif0_0)):
    a_ = np.arctan2(ydif0_0[i], xdif0_0[i]) * 180 / np.pi
    ang0_0.append(a_)

ang0_25 = []
for i in range(len(ydif0_25)):
    a_ = np.arctan2(ydif0_25[i], xdif0_25[i]) * 180 / np.pi
    ang0_25.append(a_)
    
ang0_5 = []
for i in range(len(ydif0_5)):
    a_ = np.arctan2(ydif0_5[i], xdif0_5[i]) * 180 / np.pi
    ang0_5.append(a_)
    
ang1_0 = []
for i in range(len(ydif1_0)):
    a_ = np.arctan2(ydif1_0[i], xdif1_0[i]) * 180 / np.pi
    ang1_0.append(a_)
```
